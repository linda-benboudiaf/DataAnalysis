```python
import pandas as pd
import matplotlib.pyplot as plt
unrate = pd.read_csv('./datasets/unrate.csv')
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
unrate.info()
```

```python
# Unrate employment in the USA in 1948
plt.plot(unrate['DATE'][0:12], unrate['UNRATE'][0:12])
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.xlabel('Mois')
plt.ylabel('Taux de chÃ´mage (%)')
plt.title('Evolution de taux de chÃ´mage en 1948 aux USA')
plt.show()
```

```python
fig = plt.figure(figsize=(12,12))
for i in range(5):
    ax = fig.add_subplot(5, 1, i+1)
    start_index = i*12
    end_index= (i+1)*12
    ax.plot(unrate[start_index:end_index]['DATE'], unrate[start_index:end_index]['UNRATE'])
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month

plt.figure(figsize=(6,3))
plt.plot(unrate['MONTH'][0:12], unrate['UNRATE'][0:12], c='red')
plt.plot(unrate['MONTH'][12:24], unrate['UNRATE'][12:24], c='blue')
plt.show()
```

```python
plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
#years = ['1948', '1949', '1950', '1951', '1952']
for i in range(5):
    start_index=i*12
    end_index=(i+1)*12
    label = str(1948 + i)
    plt.plot(unrate['MONTH'][start_index:end_index], unrate['UNRATE'][start_index:end_index], c=colors[i], label=label)
plt.legend(loc='upper left')
plt.title('Evolution taux de chÃ´mage aux USA de 1948 Ã  1952')
plt.xlabel('Mois (1 Ã  12)')
plt.ylabel('Taux de chÃ´mage en (%)')
plt.show()
```
